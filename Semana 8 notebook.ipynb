# Descripción del Problema
*Contexto:* Una tienda de tecnología necesita migrar su gestión de inventario de una lista lineal a una estructura jerárquica para mejorar la velocidad de consulta y organización de sus computadores. El sistema utiliza códigos numéricos únicos como clave principal para cada equipo.
Tenemos como reglas: 
1. El inventario debe construirse como un árbol binario de búsqueda (BST), donde cada nodo padre organiza a sus hijos según su valor numérico.
2. El sistema debe ser capaz de localizar códigos existentes o identificar ausencias de manera ágil.
3. Se requiere realizar recorridos (Preorden, Inorden, Postorden) para reportar el estado del inventario.
4. El sistema debe detectar cuando una inserción masiva de códigos degrada el rendimiento del árbol, al desbalancearse, y proponer el uso de árboles AVL con rotaciones para recuperar el equilibrio.

# Analisis de complejidad 
En este laboratorio se implementa un Árbol Binario de Búsqueda (BST) para gestionar los códigos de los computadores en el inventario. La eficiencia del sistema depende directamente de cómo crece este árbol, ya que todas las operaciones principales (insertar, buscar y recorrer) se realizan navegando desde la raíz hacia los nodos hijos.

Primero, es importante entender que un BST organiza los datos de forma que los valores menores van a la izquierda y los mayores a la derecha. Esto permite reducir la cantidad de comparaciones necesarias al momento de buscar o insertar un elemento, pero solo si el árbol mantiene una forma adecuada.

Las funciones más relevantes en términos de complejidad son:

- insertar(): se encarga de ubicar un nuevo código en la posición correcta del árbol.

- buscar(): permite verificar si un código ya existe o encontrar su nivel.

- recorridos (preorden, inorden, postorden): visitan todos los nodos del árbol para mostrar el inventario.

## Mejor caso: O(n log n
El mejor escenario ocurre cuando el árbol se mantiene balanceado, es decir, cuando los datos se insertan de forma que el árbol crece de manera equilibrada. En este caso, la altura del árbol es baja, aproximadamente log(n).

Esto implica que cada operación de inserción o búsqueda recorre pocos niveles del árbol, específicamente en orden logarítmico. Por lo tanto, insertar n elementos toma:

[O(n \log n)]

Este es el comportamiento ideal del sistema, ya que permite manejar grandes cantidades de datos de forma eficiente.)


```python
from IPython.display import Image
Image("peorcaso.jpeg")
```

## Peor caso 
El peor caso se presenta cuando los datos se ingresan en orden (por ejemplo: 10, 20, 30, 40...). En esta situación, el árbol deja de comportarse como un árbol y se convierte prácticamente en una lista enlazada.

Aquí la altura del árbol crece hasta n, lo que significa que cada inserción debe recorrer todos los nodos anteriores. Es decir, cada operación cuesta O(n), y al repetir esto n veces, el costo total se vuelve:

O(n^2)

Esto afecta significativamente el rendimiento del sistema, haciéndolo mucho más lento.

```python
from IPython.display import Image
Image("Grafica_pc.png")
```

El uso de un BST mejora la eficiencia en la búsqueda de datos en comparación con estructuras lineales como pilas y colas. Sin embargo, su rendimiento depende fuertemente del orden en que se insertan los elementos.

Si el árbol se mantiene balanceado, el sistema es eficiente y escalable. Pero si no se controla su forma, puede degradarse hasta un comportamiento cuadrático O(n²), perdiendo sus ventajas.

Por esta razón, en aplicaciones reales, se suelen utilizar variantes balanceadas como los árboles AVL, que garantizan un mejor rendimiento en todos los casos.

# Código


```python
# CREAR NODO

def crear_nodo(valor):
    return {
        "valor": valor,
        "izq": None,
        "der": None
    }


# INSERTAR EN BST
def insertar(inventario, valor):

    if inventario is None:
        return crear_nodo(valor)
    
    if valor < inventario["valor"]:
        inventario["izq"] = insertar(inventario["izq"], valor)

    elif valor > inventario["valor"]:
        inventario["der"] = insertar(inventario["der"], valor)

    return inventario


# RECORRIDOS DEL inventario

def preorden(inventario):

    if inventario is not None:
        print(inventario["valor"], end=" ")
        preorden(inventario["izq"])
        preorden(inventario["der"])


def inorden(inventario):

    if inventario is not None:
        inorden(inventario["izq"])
        print(inventario["valor"], end=" ")
        inorden(inventario["der"])


def postorden(inventario):

    if inventario is not None:
        postorden(inventario["izq"])
        postorden(inventario["der"])
        print(inventario["valor"], end=" ")


# DIAGRAMA DEL inventario

def grafico(raiz):

    # Si el árbol está vacío
    if raiz is None:
        print("(arbol vacio)")
        return

    # Función recursiva para dibujar hijos
    def dibujar_hijos(nodo_actual, prefijo=""):

        hijos = []

        # Agrega hijo izquierdo si existe
        if nodo_actual["izq"] is not None:
            hijos.append(("L", nodo_actual["izq"]))

        # Agrega hijo derecho si existe
        if nodo_actual["der"] is not None:
            hijos.append(("R", nodo_actual["der"]))

        # Recorre los hijos
        for indice, (lado, hijo) in enumerate(hijos):

            es_ultimo = indice == len(hijos) - 1

            # Define el conector visual
            conector = "`-- " if es_ultimo else "|-- "

            # Imprime el hijo
            print(prefijo + conector + f"{lado}: [{hijo['valor']}]")

            # Construye el nuevo prefijo para niveles inferiores
            siguiente_prefijo = prefijo + ("    " if es_ultimo else "|   ")

            # Llamada recursiva
            dibujar_hijos(hijo, siguiente_prefijo)

    # Imprime la raíz
    print(f"[{raiz['valor']}]")

    # Dibuja el resto del árbol
    dibujar_hijos(raiz)

#BUSCAR CODIGO EN EL INVENTARIO

def buscar(inventario, codigo, nivel=0):

    # Caso base: si el nodo es None, el código no existe
    if inventario is None:
        return -1   # Retorna -1 indicando que no se encontró

    # Si el valor del nodo actual es el que buscamos
    if inventario["valor"] == codigo:
        return nivel   # Retorna el nivel donde se encontró

    # Si el código es menor, se busca en el subárbol izquierdo
    if codigo < inventario["valor"]:
        return buscar(inventario["izq"], codigo, nivel + 1)

    # Si el código es mayor, se busca en el subárbol derecho
    else:
        return buscar(inventario["der"], codigo, nivel + 1)


# MENU PARA USUARIO

raiz = None

while True:

    print("\n====== INVENTARIO DE COMPUTADORES ======")
    print("1. Agregar código del objeto al inventario.")
    print("2. Visualizar el inventario.")
    print("3. Buscar codigo.")
    print("4. Terminar programa.")

    opcion = input("Seleccione una opcion (1, 2, 3 o 4): ")

    
    # AGREGAR CODIGO
    
    if opcion == "1":

        entrada = input("Ingrese código del computador (solo numeros un codigo a la vez): ")
        #Verifica si es un numero
        try:
            codigo = int(entrada)

            #verificar si es mayor a 0
            if codigo>=0:

                # Verifica si el código ya existe
                if buscar(raiz, codigo) != -1:
                    print("El codigo ya existe en el inventario, no se puede repetir")

                else:
                    raiz = insertar(raiz, codigo)
                    print("Codigo agregado correctamente")

            else:
                print("Codigo invalido: ingrese numeros enteros positivos")

        #si son caracteres diferentes a numeros
        except ValueError:
            print("Error: debe ingresar solo numeros enteros positivos")


    
    # VER INVENTARIO 
    
    elif opcion == "2":

        if raiz is None:

            print("El inventario esta vacio")

        else:

            print("\n--- RECORRIDOS ---")

            print("Preorden:")
            preorden(raiz)

            print("\nInorden:")
            inorden(raiz)

            print("\nPostorden:")
            postorden(raiz)

            print("\n\n\t DIAGRAMA DEL INVENTARIO ")
            print("Leer de izquierda a derecha.")

            grafico(raiz)
    
    elif opcion == "3":

        # Verifica si el árbol está vacío
        if raiz is None:
            print("El inventario esta vacio")

        else:
             # Solicita el código al usuario
            entrada = input("Ingrese el codigo a buscar: ")

            try:
                # Convierte la entrada a número entero
                codigo = int(entrada)

                # Llama a la función buscar
                nivel = buscar(raiz, codigo)

                # Si el nivel es diferente de -1, el código fue encontrado
                if nivel != -1:
                    print("El codigo SI se encuentra en el inventario, en el nivel" ,nivel)

                else:
                    # Si retorna -1, no se encontró
                    print("El codigo NO se encuentra en el inventario")

            except ValueError:
                # Maneja el error si el usuario no ingresa un número
                print("Error: debe ingresar solo numeros enteros")



    
    # TERMINAR PROGRAMA
    
    elif opcion == "4":

        print("Programa terminado.")
        break


    else:

        print("Opcion invalida, debe ingresa 1, 2 o 3.")
```
